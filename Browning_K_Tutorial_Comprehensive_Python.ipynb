## Another class with explanation

```python

#An example of a class
class Box:

    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.description = "awaiting description of box"
        self.author = "awaiting author of box"

    def area(self):
        return self.x * self.y

    def perimeter(self):
        return 2 * self.x + 2 * self.y

    def describe(self, text):
        self.description = text

    def authorName(self, text):
        self.author = text

    def scaleSize(self, scale):
        self.x = self.x * scale
        self.y = self.y * scale
```

This class or blueprint gives us a  description of a box.This is not the actual box just the description of what a box is. The box has a width (x), a height (y), and an area and perimeter (area(self) and perimeter(self)). We are not running code, just defining the class and it's variables and functions

A function called __init__ is designed to run at the creation of an instance of the box.  

self is how we refer to things in the class. self is the first parameter in any function defined inside a class. Any function or variable created on the first level of indentation (that is, lines of code that start one TAB to the right of where we put class Box) is automatically put into self. 

To use a class you need to create an instance.  The code below defines the instance of Box and it's parameters. 

```python
rectangle = Box(120, 60)
```

Once this code is ran the __init__ function comes into play. It creates an instance of a class Box and uses the values in the brackets to satisfy the parameters of self.  This box is a rectangle

The rectangle is an instance of class Box.  

```python
#find the area of rectangle:
print(rectangle.area())

#find the perimeter of rectangle:
print(rectangle.perimeter())

#describe the rectangle
rectangle.describe("This is a rectangle that is wider than it is tall")

```
